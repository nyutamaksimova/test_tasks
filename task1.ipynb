```python
df_p = pd.DataFrame(df.groupby(['pub'], as_index = False)['revenue'].mean().sort_values(ascending = False, ignore_index=True, by=['revenue'])['pub'])
```

```python
df_w = pd.DataFrame(df.groupby(['worker'], as_index = False)['revenue'].mean().sort_values(ascending = False,ignore_index=True, by=['revenue'])['worker'])
```

```python
result = df_p.join(df_w)
```
